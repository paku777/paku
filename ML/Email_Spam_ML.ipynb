```python
import pandas as pd

df=pd.read_csv("/content/emails.csv")
df.head()

df.describe()

df.isnull().sum()


df=df.dropna()


df.isnull().sum()


X=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)


import numpy as np
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)



from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)


prediction=knn.predict(X_test)
prediction



from sklearn.metrics import confusion_matrix
conf_matrix=confusion_matrix(prediction,y_test)
conf_matrix



confusion_df=pd.DataFrame(conf_matrix,index=['Actual 0','Actual 1'],columns=['predicted 0','predicted 1'])
confusion_df



import seaborn as sns
sns.heatmap(confusion_df,cmap='coolwarm',annot=True)




from sklearn.metrics import r2_score,mean_squared_error,accuracy_score,classification_report
print("R2 SCORE",r2_score(prediction,y_test))
print("MEAN_SQUARED_ERROR",mean_squared_error(prediction,y_test))
print("ACCURACY SCORE",accuracy_score(prediction,y_test))



from sklearn import metrics
print("\n\n*** CLASSIFICATION REPORT ***\t\n",metrics.classification_report(prediction,y_test))



from sklearn.svm import SVC
svc=SVC(kernel='rbf',random_state=0)
svc.fit(X_train,y_train)
predictions=svc.predict(X_test)
predictions



from sklearn.metrics import r2_score,mean_squared_error,accuracy_score,confusion_matrix,classification_report
print("R2 SCORE",r2_score(predictions,y_test))
print("MEAN_SQUARED_ERROR",mean_squared_error(predictions,y_test))
print("ACCURACY SCORE",accuracy_score(predictions,y_test))




conf_mat=confusion_matrix(y_test,predictions)
conf_mat



print("\n\n*** CLASSIFICATION REPORT ***\t\n",classification_report(predictions,y_test))



conf_df=pd.DataFrame(conf_mat,index=['Actual 0','Actual 1'],columns=['Predicted 0','Predicted 1'])
conf_df
sns.heatmap(conf_df,cmap='coolwarm',annot=True)

```
